### Creating connection with the project's database and creating the tables in it

```python
connection_string = f"{driver}//{user}:{password}@{ip}/{database}"
engine = create_engine(connection_string)
```

```python
query_country = """
CREATE TABLE country(
id INT(17) NOT NULL AUTO_INCREMENT,
country VARCHAR(25),

PRIMARY KEY(id)
);
"""
```

```python
engine.execute(query_country)
```

```python
query_elections = """
CREATE TABLE elections(
id INT(17) NOT NULL AUTO_INCREMENT,
year INT,
id_country INT(17),
proportion_votes FLOAT,
extreme_type VARCHAR(30),

PRIMARY KEY(id),
FOREIGN KEY(id_country) REFERENCES country(id)
);
"""

```

```python
engine.execute(query_elections)
```
